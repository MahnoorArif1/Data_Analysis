The Lowest Starting and Mid-Career Salary

```python
print('The Lowest Starting Salary')
print(clean_data['Undergraduate Major'].loc[clean_data['Starting Median Salary'].idxmin()])
print('The Lowest Mid-Career Salary')
print(clean_data['Undergraduate Major'].loc[clean_data['Mid-Career Median Salary'].idxmin()])
```

```python
spread_col = clean_data['Mid-Career 90th Percentile Salary'] - clean_data['Mid-Career 10th Percentile Salary']
clean_data.insert(1, 'Spread', spread_col)
clean_data.head()
```

```python
print(clean_data['Undergraduate Major'].loc[clean_data['Spread'].idxmin()])
```

```python
sorted_clean_data=clean_data.sort_values('Mid-Career 90th Percentile Salary')
sorted_clean_data['Undergraduate Major'].head()
```

```python
sorted_clean_data=clean_data.sort_values('Spread')
sorted_clean_data[['Undergraduate Major','Spread']].head()
```
